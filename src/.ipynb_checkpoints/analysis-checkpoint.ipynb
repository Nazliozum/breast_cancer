# Data

The data comes from the [UCI Machine Learning Repository](http://archive.ics.uci.edu/ml/index.php) and links features computed from a digitized image of a fine needle aspirate (FNA) of a breast mass to whether the mass is malignant or benign. The dataset can be downloaded [here](http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data).

- There are 32 attributes in the dataset including **ID**, **diagnosis** (target variable) and **30 real-valued features** (input variables).


- Target variable has 2 values: 
    - M = malignant
    - B = benign


- The input variables consist of the mean, standard error, and "worst" or largest (mean of the three largest values) of the 10 characteristics of the cell nuclei present in each image, resulting in 30 features in total. The 10 characteristics are as follows:

```
    a) radius (mean of distances from center to points on the perimeter)
	b) texture (standard deviation of gray-scale values)
	c) perimeter
	d) area
	e) smoothness (local variation in radius lengths)
	f) compactness (perimeter^2 / area - 1.0)
	g) concavity (severity of concave portions of the contour)
	h) concave points (number of concave portions of the contour)
	i) symmetry 
	j) fractal dimension ("coastline approximation" - 1)
```

- The dataset consists of 569 observations:
    - 357 belonging to benign class
    - 212 belonging to malignant class.
    
- There are no missing values in the dataset.

More information on the dataset: [Wisconsin Diagnostic Breast Cancer (WDBC)](http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.names)

# Dependencies

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import neighbors, svm
from sklearn.model_selection import train_test_split, GridSearchCV, ShuffleSplit
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
````

# Import data

````python
# Import data
cancer = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data", header = None)

# Write the data frame to a csv file
cancer.to_csv("../data/wdbc.csv", encoding='utf-8', index=False, header=False)
````

# Exploratory Data Analysis

````python
cancer.head()
````

The data table does not say us much as it is. It is better to include the column names to have a better idea.

````python
cancer = cancer.rename(index=str, columns={0: "id", 1: "target", 2: "mean radius", 3: "mean texture", 4: "mean perimeter", 5: "mean area",
                                  6: "smoothness", 7: "mean compactness", 8: "mean concavity", 9: "mean concave points", 
                                  10: "mean symmetry", 11: "mean fractal dimension", 12: "radius error", 13: "texture error", 
                                  14: "perimeter error", 15: "area error", 16: "smoothness error", 17: "compactness error", 
                                  18: "concavity error", 19: "concave points error", 20: "symmetry error", 21: "fractal dimension error", 
                                  22: "worst radius", 23: "worst texture", 24: "worst perimeter", 25: "worst area", 26: "worst smoothness", 
                                  27: "worst compactness", 28: "worst concavity", 29: "worst concave points", 30: "worst symmetry", 
                                  31: "worst fractal dimension"})
````

````python
cancer.head()
````

Below is a collection of histogram plots for each of the features. Each histogram shows the count of how often an observation appears having the feature in a certain range. These ranges appear as bins on the plots. The data is divided into 2 groups according to the target variable: malignant and benign. Therefore, if we see a bin coloured only in red, then this means that only observations in the benign category have values for that specific feature in that range.

````python
fig, axes = plt.subplots(15, 2, figsize=(10, 20))
malignant = (cancer.loc[cancer['target'] == 'M'])
benign = cancer.loc[cancer['target'] == 'B']

ax = axes.ravel()

for i in range(2,32):
    _, bins = np.histogram(cancer.iloc[:, i], bins=50)
    ax[i-2].hist(malignant.iloc[:, i], bins=bins, color = 'red', alpha=.5)
    ax[i-2].hist(benign.iloc[:, i], bins=bins, color = 'green', alpha=.5)
    ax[i-2].set_title(cancer.columns[i])
    ax[i-2].set_yticks(())

ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["malignant", "benign"])    
fig.tight_layout()
````

The histograms show us some valueable information as to which features are better at distinguishing a malignant tumor from a benign one. For example, features such as `smoothness error`, `texture error` or `mean fractal dimension` give us little idea since observations from both categories overlap in the same regions. However, `mean radius` and `mean perimeter` demonstrate clear differences between malignant and benign tumors which proves them as important features to be included in a prediction model.

Since we have realized that some features are not so important in determining the characteristic of the breast mass, we can have the idea to eliminate these features. However, it is good to take a look at the correlations between features too. Below is another plot that inspects the correlation among features.

````python
corr = cancer.iloc[:,1:32].corr()

sns.set(style="white")

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````

We can generally observe correlation, either positive or negative, between most of the features.

# Model Fitting

There are many machine learning models that we can fit to this dataset. However, following the [cheatsheet](http://peekaboo-vision.blogspot.ca/2013/01/machine-learning-cheat-sheet-for-scikit.html) prepared by [Andreas Mueller](http://peekaboo-vision.blogspot.ca/), I have decided to carry on with [Linear SVC](http://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVC.html#sklearn.svm.LinearSVC) because the breast cancer dataset is a labelled classification problem with 569 samples which falls into the `<100K samples` category.

![](http://1.bp.blogspot.com/-ME24ePzpzIM/UQLWTwurfXI/AAAAAAAAANw/W3EETIroA80/s1600/drop_shadows_background.png)

Now, the next step is to fit the model and have a look at its the performance.

````python
# Separate attributes and labels in the dataset.
X = np.array(cancer.iloc[:,2:32])
y = np.array(cancer.iloc[:,1])
````

````python
# Split the data as training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1992)
````

````python
print(X_train.shape)
print(X_test.shape)
````

We should not forget to scale the data since we are using an SVM classifier.

````python
# preprocessing using 0-1 scaling
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

The `LinearSVM` algorithm uses `L2` regularization by default. Therefore, we will not be doing any extra feature selection since all the features seemed to be correlated to each other in some way and they all seem to be effective. Another important decision is to determine the `C` parameter in the algorithm.

````python
parameters = {'C':[10**-4, 10**-3, 10**-2, 0.1, 1, 10, 10**2, 10**3, 10**4]}

svc = svm.LinearSVC()

cv = ShuffleSplit(n_splits=5, test_size=0.25, random_state=0)
clf = GridSearchCV(svc, parameters, cv=cv)
clf.fit(X_train_scaled, y_train)
print("Optimal C for LinearSVM: {}".format(clf.best_params_.get('C')))
print("Maximum accuracy score of {:,.4f} reached with C = {}".
      format(clf.best_score_, clf.best_params_.get('C')))
````

The cross-validation provides us an unbiased approximation of the test accuracy and shows that using `C = 1` will give us the highest test score. 

````python
# Fit the model
svc = svm.LinearSVC(C=10, random_state=0)
svc.fit(X_train_scaled, y_train)
````

# Final Step: Test Accuracy

````python
# scoring on the scaled test set
print("Test accuracy: {:.2f}".format(
    svc.score(X_test_scaled, y_test)))
````
